Part 3: Modifying Data 

```python
# Create a new column called Performance using the following rule: 
# • If GPA >= 3.5 → "Excellent" 
# • Otherwise → "Good" 
df["Performance"] = df["GPA"].apply(lambda x: "Excellent" if x >= 3.5 else "Good")
df

```

Part 5: Combining Data 

```python
# Create a new DataFrame called activities with the following data: 
# StudentID  
# Scholarship Club 
# 1 Yes Robotics 
# 2 No Sports 
# 3 Yes Music 
# 4 No Art 
# 5 Yes Coding 
 
 
# Then: 
# 1. Save it as activities.csv. 
# 2. Read the file into a new DataFrame. 
# 3. Display the first 5 rows
data = [
    {"StudentID": 1, "Scholarship": "Yes", "Club": "Robotics"},
    {"StudentID": 2, "Scholarship": "No",  "Club": "Sports"},
    {"StudentID": 3, "Scholarship": "Yes", "Club": "Music"},
    {"StudentID": 4, "Scholarship": "No",  "Club": "Art"},
    {"StudentID": 5, "Scholarship": "Yes", "Club": "Coding"}
]
activities = pd.DataFrame(data)
activities
activities.to_csv("activities.csv", index=False)
df_activities=pd.read_csv("activities.csv")
df_activities.iloc[0:5]
```

```python
# Merge the original student dataset with activities.csv using the StudentID column. 
# Display the merged DataFrame

merged_df=pd.merge(df,df_activities, on="StudentID")
merged_df
```

```python
# Create another DataFrame containing 3 new students with the same columns as the 
# original dataset. 
# Then: 
# 1. Save it as new_students.csv. 
# 2. Read the file. 
# 3. Append it to the original dataset using concat(). 
# 4. Display the last 6 rows of the final DataFrame.

new_students = pd.DataFrame({
    "StudentID": [11, 12, 13],
    "Gender": ["Male", "Female", "Male"],
    "Age": [20, 22, 21],
    "StudyHoursPerWeek": [25, 30, 18],
    "AttendanceRate": [90.5, 85.0, 78.3],
    "GPA": [3.6, 3.2, 2.9],
    "Major": ["Engineering", "Science", "Business"],
    "PartTimeJob": ["No", "Yes", "No"],
    "ExtraCurricularActivities": ["Yes", "No", "Yes"],
    "Performance": ["Excellent", "Good", "Good"]
})
new_students
new_students.to_csv("new_students.csv", index=False)
pd.read_csv("new_students.csv")

final_df = pd.concat([df, new_students], ignore_index=True)
final_df.tail(6)
```
